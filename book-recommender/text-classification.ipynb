```python
category_mapping = {'Fiction': 'Fiction',
                    'Juvenile Fiction': "Children's Fiction",
                    'Biography & Autobiography': 'Nonfiction',
                    'History': 'Nonfiction',
                    'Literary Criticism': 'Nonfiction',
                    'Philosophy': 'Nonfiction',
                    'Religion': 'Nonfiction',
                    'Comics & Graphic Novels': 'Fiction',
                    'Drama': 'Fiction',
                    'Juvenile Nonfiction': 'Nonfiction',
                    'Science': 'Nonfiction',
                    'Poetry': 'Fiction'}
```

```python
books['simplified_categories'] = books.categories.map(category_mapping)
```

```python
books
```

```python
books[~(books['simplified_categories'].isna())]
```

```python
from transformers import pipeline
categories = ["Fiction", "Nonfiction"]
pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device="mps")
```

```python
#!pip install tf-keras
```

```python
sequence = books.loc[books['simplified_categories'] == 'Fiction', "description"].reset_index(drop=True)[0]
```

```python
sequence
```

```python
pipe(sequence, categories)
```

```python
import numpy as np
def generate_predictions(sequence, categories):
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions['scores'])
    return predictions['labels'][max_index]
```

```python
generate_predictions(sequence, categories)
```

```python
from tqdm import tqdm

actual_cats = []
predicted_cats = []

for i in tqdm(range(0, 300)):
    sequence = books.loc[books['simplified_categories'] == 'Fiction', "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, categories)]
    actual_cats += ["Fiction"]

for i in tqdm(range(0, 300)):
    sequence = books.loc[books['simplified_categories'] == 'Nonfiction', "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, categories)]
    actual_cats += ["Nonfiction"]
```

```python
predictions_df = pd.DataFrame({'actual_categories': actual_cats, 'predicted_categories': predicted_cats})
```

```python
predictions_df.head()
```

```python
predictions_df['correct_prediction'] = np.where(predictions_df['actual_categories'] == predictions_df['predicted_categories'], 1, 0)
```

```python
predictions_df['correct_prediction'].sum() / len(predictions_df)
```

```python
isbns = []
predicted_cats = []
missing_cats = books.loc[books['simplified_categories'].isna(), ["isbn13", "description"]].reset_index(drop=True)
for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats["description"][i]
    predicted_cats += [generate_predictions(sequence, categories)]
    isbns += [missing_cats["isbn13"][i]]
```

```python
missing_predicted_df = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_cats})
```

```python
missing_predicted_df
```

```python
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")
books['simple_categories'] = np.where(books['simplified_categories'].isna(), books['predicted_categories'], books['simplified_categories'])
```

```python
books = books.drop(columns=['predicted_categories'])
```

```python
books.to_csv("books_with_categories.csv", index=False)
```
